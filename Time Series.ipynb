# Time resampleing

```python
tesla.head()
```

```python
tesla.resample(rule="A").min()
```

```python
tesla.resample(rule="A").max()
```

```python
tesla.resample(rule="A").max()["Open"].plot()
```

```python
#Quaterly start frequency
tesla.resample(rule="QS").max()["High"].plot()
```

```python
#business end frequency
tesla.resample(rule="BA").max()["High"].plot()
```

```python
#business end frequency
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
tesla.resample(rule="BQS").max()
```

```python
#ploting
tesla["Open"].resample(rule="A").mean().plot(kind="bar")
```

```python
tesla["Open"].resample(rule="M").mean()
```

```python
tesla["Open"].resample(rule="A").max()
```

```python
tesla["Open"].resample(rule="M").max().plot(kind="bar", figsize=(30,15))
```

```python
tesla["High"].rolling(11).mean().head(20)
```

```python
tesla["Open: 30 days rolling"] = tesla["Open"].rolling(30).mean()
```

```python
tesla.head(40)
```

```python
tesla[["Open", "Open: 30 days rolling"]].plot(figsize=(12,8))
```
